```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Create a random dataset that has a normal distribution and then sort it (in this case, 1000 data points)
x = np.random.normal(50,25,5000)
x = np.sort(x)

#Create another dataset to put on the y axis of the scatter plot
y = np.random.normal(50,25,5000)

#plot the dataset
plt.scatter(x,y,alpha=0.3)
plt.show()
```

```python
df = pd.DataFrame({'x':x, 'y':y})
df.to_csv("C:/temp/pbitest5000.csv")
```
